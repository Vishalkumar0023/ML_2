# Unit 3: Unsupervised Learning - K-Means Clustering

## Theory

### Algorithm: K-Means Clustering

**How it works:**
- K-Means groups data into K clusters.
- It assigns points to nearest centroid and updates centroids iteratively.

**Why selected:**
- Simple and fast
- Useful for grouping similar patients
- Works well on numerical datasets

### Supervised vs Unsupervised Learning

| Supervised | Unsupervised |
|------------|-------------|
| Uses labeled data | Uses unlabeled data |
| Predicts output | Finds patterns |
| Example: Logistic Regression | Example: K-Means |


```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load dataset
df = pd.read_csv("cardio.csv")

# Data cleaning
df = df.drop_duplicates()
df = df.drop('id', axis=1)
df['age'] = df['age'] / 365
df = df.fillna(df.mean(numeric_only=True))

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

print(df[['Cluster']].head())

# Visualization
plt.figure()
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=df['Cluster'])
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
